# Question 1.

Your Friend has developed the Product and he wants to establish the product startup and he is searching for a perfect location where getting the investment has a high chance. But due to its financial restriction, he can choose only between three locations -  Bangalore, Mumbai, and NCR. As a friend, you want to help your friend deciding the location. NCR include Gurgaon, Noida and New Delhi. Find the location where the most number of funding is done. That means, find the location where startups has received funding maximum number of times. Plot the bar graph between location and number of funding. Take city name "Delhi" as "New Delhi". Check the case-sensitiveness of cities also. That means, at some place instead of "Bangalore", "bangalore" is given. Take city name as "Bangalore". For few startups multiple locations are given, one Indian and one Foreign. Consider the startup if any one of the city lies in given locations.

```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
file_open=open('startup_funding.csv')
```

```python
file_orignal=pd.read_csv(file_open)
file=file_orignal.copy()
```

```python
file.dropna(subset=['CityLocation'],inplace=True)
file.reset_index(inplace=True,drop=True)
```

```python
file.loc[(file['CityLocation']=='Gurgaon')|(file['CityLocation']=='Noida ')|(file['CityLocation']=='New Delhi'),'CityLocation']='NCR'
```

```python
file.replace('bangalore','Bangalore',inplace=True)
file.replace('Delhi','New Delhi',inplace=True)
```

```python
def update_function(city):
    return city.split('/')[0].strip()
```

```python
file['CityLocation'].apply(update_function)
```

```python
new_file=file[(file['CityLocation']=='Bangalore')|(file['CityLocation']=='Mumbai')|(file['CityLocation']=='NCR')]
```

```python
city_count=new_file['CityLocation'].value_counts()
city=city_count.index
count=city_count.values

print(city)
print(count)
li = list(zip(city, count))

#build pandas dataframe 
freq = pd.DataFrame(li, columns = ["Name Of City", "Number Of Fundings"])
freq
```

```python
plt.bar(city,count)
```

```python
plt.title('cities vs no of funding')
color=['blue','red','purple']
plt.bar(city,count,color=color,width=0.9)
plt.xlabel('name of city')
plt.ylabel('no.of company give funding')

plt.show()
```

```python
plt.title('no.of company vs city')
explode=[0.1,0,0]
plt.pie(count,autopct='%.2f%%',labels=city,explode=explode)
plt.axis('equal')

plt.show
```
