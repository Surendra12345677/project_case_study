```python
import numpy as np
```

```python
list=[1,2,3,4]
#mandatry arument list in array
numpy_array=np.array(list)
print(numpy_array)
print(type(numpy_array))
```

```python
#numpy array only contain homogenous element example:
list=[1,2,3,4,'a',0.5]
numpy_array=np.array(list)
print(numpy_array)
#iven below all are type of string
```

```python
#you can type cast using one more arument:
list=[1,2,3,4,0.4,0.3]
numpy_array_new=np.array(list,dtype=int)
print(numpy_array_new)
```

```python
list=[1,2,34,5]
numpy_array=np.array(list*3)
print(numpy_array)
```
